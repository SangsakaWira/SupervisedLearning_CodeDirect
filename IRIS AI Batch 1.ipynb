```
import keras
import numpy as np
import pandas as pd
import seaborn as sns
from keras import Sequential
from keras.layers import Dense
```

```
dataset = pd.read_csv("iris.csv")
```

```
dataset
```

```
x = dataset.iloc[:,:-1].values
y = pd.get_dummies(dataset["variety"]).values
```

```
y
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```
y_test
```

```
X_test.shape
```

```
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu', input_dim = 4))

# Adding the second hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the second hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the second hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the second hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the second hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 3, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100, validation_data=(X_test,y_test))
```

```
hasil = classifier.predict(X_test)
```

```
hasil = np.round(hasil)
test = y_test
validasi = pd.DataFrame()
validasi["Setosa"] = test.T[0]
validasi["Versicolor"] = test.T[1]
validasi["Virginica"] = test.T[2]
validasi["Pred Setosa"] = hasil.T[0]
validasi["Pred Versicolor"] = hasil.T[1]
validasi["Pred Virginica"] = hasil.T[2]
```
