```python
new_df = df.copy()
holiday_one_hot = pd.get_dummies(df['holiday'])
new_df = new_df.drop(columns = 'holiday')
new_df = new_df.join(holiday_one_hot)
weather_m_one_hot = pd.get_dummies(df['weather_main'])
new_df = new_df.drop(columns = 'weather_main')
new_df = new_df.join(weather_m_one_hot)
weather_desc_one_hot = pd.get_dummies(df['weather_description'])
new_df = new_df.drop(columns = 'weather_description')
new_df = new_df.join(weather_desc_one_hot)
```

```python
new_df
```

```python
new_df['year'] = df['date_time'].str.split('-').str.get(0)
new_df['month'] = df['date_time'].str.split('-').str.get(1)
new_df['day'] = df['date_time'].str.split('-').str.get(2).str.split().str.get(0)
new_df['hour'] = df['date_time'].str.split().str.get(1).str.split(':').str.get(0)
```

```python
new_df = new_df.drop(columns = 'date_time')
```

```python
new_df
```

```python
y = new_df['traffic_volume'].values
x = new_df.drop(columns = 'traffic_volume')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=32)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = keras.Sequential([
keras.layers.Dense(1000, activation=tf.nn.relu, input_shape=[69]),
keras.layers.Dense(1000, activation=tf.nn.relu),
keras.layers.Dense(700, activation=tf.nn.relu),
keras.layers.Dense(600, activation=tf.nn.relu),
keras.layers.Dense(5000, activation=tf.nn.relu),
keras.layers.Dense(1, activation=tf.nn.relu)
])
model.compile(optimizer='adam',loss='mean_absolute_error')
save_best_val_loss = keras.callbacks.ModelCheckpoint('best_model_val_loss.h5',save_best_only=True,monitor='val_loss',mode='min')
save_best_loss = keras.callbacks.ModelCheckpoint('best_model_loss.h5',save_best_only=True,monitor='loss',mode='min')
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test, y_test),epochs = 100,batch_size = 400,shuffle = True,callbacks = [save_best_val_loss, save_best_loss])
```
